```python
data1 = pd.read_excel('cra/조회순.xlsx')
data2 = pd.read_excel('cra/thank조회순.xlsx')
```

```python
data1['조회수'] = data1['조회수'].str.replace('조회수 ','')
data1['제목'] = data1['제목'].apply(lambda x : x.split('] ')[1])
```

```python
result = '\n'.join(s for s in data1['아이콘'])
result = result.replace('\n',',')
l = result.split(sep=',')
s = set(l)-set([''])
```

```python
for i in s:
    data1_s1 =  data1[data1['아이콘'].str.contains(i)]
    data1_s2 = data1[~data1['아이콘'].str.contains(i)]
    data1_s1[i] = 1
    data1_s2[i] = 0
    data1 = pd.concat([data1_s1,data1_s2])
```

```python
data1 = data1.drop('아이콘',axis=1)
```

```python
data1.to_csv('크롤링전처리완료.csv',encoding='UTF-8')
```
