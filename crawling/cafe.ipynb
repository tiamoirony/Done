```python
page = int(input("페이지입력"))

wb = openpyxl.Workbook()
ws = wb.active 


ws['A1'] = '캠핑장 정보'
ws.append(['캠핑장','date'])


for num in range(127, 134):
    for i in range(1, page+1):

    
        driver = webdriver.Chrome('/usr/local/bin/chromedriver')
        
        addr = (f'https://cafe.naver.com/ArticleList.nhn?search.clubid=20486145&search.menuid={num}&search.boardtype=L&search.totalCount=151&search.cafeId=20486145&search.page={i}')
        
        driver.get(addr)
        
        driver.switch_to.frame('cafe_main')
        
        html = driver.page_source
        
        soup = BeautifulSoup(html, 'html.parser')
    
    
    
        infos = soup.select('tr')
        
        for info in infos:
            
            
            try:
                title = info.select('.article')[0].text.strip()
                title.replace(' ','')
                # title = title.split(' ')[1:]
                # title = title[5:22]
                # title = title.replace(' ','')
                

                datetime = info.select('.td_date')[0].text.strip()
                date = datetime[:10]
                
                print(title, date)
                
                ws.append([title, date])
                
            except:
                pass
            

wb.save("캠핑장11.xlsx") 


        
```
